Preparing Data for Convulation Layer version

```python
from random import shuffle
import glob
import sklearn
```

```python
shuffle_data = True 
```

```python
train_hdf5_path = 'spec_train.h5'  # file path for the created .hdf5 file

spec_train_path = 'pixel_train/*.png' # the original data path

test_hdf5_path = 'spec_test.h5'  # file path for the created .hdf5 file

spec_test_path = 'pixel_test/*.png' # the original data path
```

```python
# get all the image paths 
train_address = glob.glob(spec_train_path)
test_address = glob.glob(spec_test_path)

print(train_address)
```

```python
# label the data

train_labels=[]
test_labels=[]
class_info=[]

#angry, disgust, fear, happy, neutral, sad

for train_addr in train_address:
    if 'angry' in train_addr:
        train_labels.append(0)
    elif 'disgust' in train_addr:
        train_labels.append(1)
    elif 'fear' in train_addr:
        train_labels.append(2)
    elif 'happy' in train_addr:
        train_labels.append(3)
    elif 'neutral' in train_addr:
        train_labels.append(4)
    elif 'ps' in train_addr:
        train_labels.append(5)
    else: 
        train_labels.append(6)
        
for test_addr in test_address:
    if 'angry' in test_addr:
        test_labels.append(0)
        class_info.append("angry")
    elif 'disgust' in test_addr:
        test_labels.append(1)
        class_info.append("disgust")
    elif 'fear' in test_addr:
        test_labels.append(2)
        class_info.append("fear")
    elif 'happy' in test_addr:
        test_labels.append(3)
        class_info.append("happy")
    elif 'neutral' in test_addr:
        test_labels.append(4)
        class_info.append("neutral")
    elif 'ps' in test_addr:
        test_labels.append(5)
        class_info.append("ps")
    else: 
        test_labels.append(6)
        class_info.append("sad")
print(train_labels)
print(class_info)
```

```python
if shuffle_data:
    c_train = list(zip(train_address, train_labels)) 
    c_test = list(zip(test_address, test_labels, class_info)) 
    shuffle(c_train)
    shuffle(c_test)
 
    (train_address, train_labels) = zip(*c_train) 
    (test_address, test_labels, class_info) = zip(*c_test) 
```

Create the h5py project

```python
import numpy as np
import h5py
```

```python
train_shape = (len(train_address), 32, 32, 3)
test_shape = (len(test_address), 32, 32, 3)
```

```python
# open a hdf5 file and create earrays 
f_train = h5py.File(train_hdf5_path, mode='w')
f_test = h5py.File(test_hdf5_path, mode='w')
```

```python
f_train.create_dataset("train_image", train_shape, np.uint8)
f_test.create_dataset("test_image", test_shape, np.uint8)  
```

```python
dt = h5py.special_dtype(vlen=str)

f_train.create_dataset("train_labels_key", (len(train_address),), np.float16)
f_train["train_labels_key"][...] = train_labels

f_test.create_dataset("test_labels_key", (len(test_address),), np.float16)
f_test["test_labels_key"][...] = test_labels

f_test.create_dataset("classes", (len(test_address),), dt)
f_test["classes"][...] = class_info
```

Write the images

```python
import cv2
```

```python
# loop over train paths
for i in range(len(train_address)):
  
    if i % 1000 == 0 and i > 1:
        print ('Train data: {}/{}'.format(i, len(train_address)) )

    addr = train_address[i]
    img = cv2.imread(addr)
    img = cv2.resize(img, (32, 32), interpolation=cv2.INTER_CUBIC)# resize to (64*64)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # cv2 load images as BGR, convert it to RGB
    f_train["train_image"][i, ...] = img[None] 
```

```python
# loop over test paths
for i in range(len(test_address)):
  
    if i % 1000 == 0 and i > 1:
        print ('Test data: {}/{}'.format(i, len(test_address)) )

    addr = test_address[i]
    img = cv2.imread(addr)
    img = cv2.resize(img, (32, 32), interpolation=cv2.INTER_CUBIC)# resize to (128,128)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # cv2 load images as BGR, convert it to RGB
    f_test["test_image"][i, ...] = img[None] 
```

```python
f_train.close()
f_test.close()
```
